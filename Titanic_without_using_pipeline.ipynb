```python
from sklearn.impute import SimpleImputer

# Separate imputers for each column
si_age = SimpleImputer(strategy='mean')
si_Fare = SimpleImputer(strategy='most_frequent')

# Fit each imputer on its own column
si_age.fit(x_train[['Age']])
si_Fare.fit(x_train[['Fare']])

# Transform both train and test
x_train['Age'] = si_age.transform(x_train[['Age']])
x_test['Age'] = si_age.transform(x_test[['Age']])

x_train['Fare'] = si_Fare.transform(x_train[['Fare']])
x_test['Fare'] = si_Fare.transform(x_test[['Fare']])
```
